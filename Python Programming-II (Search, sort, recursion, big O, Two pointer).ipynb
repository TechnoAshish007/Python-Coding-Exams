```python
# Binary search in a list

#input
import ast
sorted_list = ast.literal_eval(input()) #input list
key = int(input()) #input key

# Task :: find the index of the key in the list

def binary_search(sorted_list, key):
    l = 0 #left start point
    r = len(sorted_list)-1 #right start point
    while (l <= r):
        mid =  (l+r)//2 #mid element

        if sorted_list[mid]== key:
            return(mid)
        elif sorted_list[mid]<key:
            l = mid+1
        else:
            r= mid-1
    return(-1)

print(binary_search(sorted_list, key))
```

```python
# Binary search in Rotated list
#input
import ast
data = ast.literal_eval(input()) #input list
key = int(input()) #input ke

def rotated_array(data, key):
    l = 0
    r = len(data)-1
    
    while (l <= r):
        mid = (l+r)//2
        if data[mid]==key:
            return mid
        elif data[l]<= data[mid]:
            if key >= data[l] and key < data[mid]:
                r = mid -1
            else:
                l = mid+1
        elif data[l]> data[mid]:
            if key > data[mid] and key <= data[r]:
                l = mid -1
            else:
                r = mid+1
    return -1
print(rotated_array(data, key))
```

```python
# Use binary search and find the square root of n.

# input 
n = int(input())

def squareroot(n):
    l = 0
    r = n
    while abs(l-r)> 10**-5:
        mid = (l+r)/2
        if mid**2 > n:
            r = mid
        else:
            l = mid
    return(round(mid, 4))
print("%.4f" % squareroot(n))
```

```python
# In the binary search, you kept reducing the search space into half of the previous search space. In this question, 
# can you similarly reduce the search space in one-third of the last space search and write a searching algorithm?
import ast
mylist = ast.literal_eval(input())
key  = int(input())
def ternary_search(mylist,key):
    l = 0
    r = len(mylist)-1
    while l<=r:
        mid1 = l + (r-l)//3
        mid2 = r - (r-l)//3
        if key == mylist[mid2]:
            return (mid2)
        elif key == mylist[mid1]:
            return (mid1)
        elif mylist[l] <= key and mylist[mid1] > key and mylist[mid2]>key:
            r = mid1 -1
        elif mylist[mid1] <= key and mylist[mid2] > key:
            l = mid1 + 1
            r = mid2 -1
        elif mylist[mid2]< key and mylist[r]>= key:
            l = mid2 + 1
        else:
            r = mid2 -1
    return (-1)

print(ternary_search(mylist,key))
```

```python
# Insertion sort
# Python program for implementation of Insertion Sort 
  
# Function to do insertion sort 
def insertionSort(arr):
    
  
    # Traverse through 1 to len(arr) 
    for i in range(1, len(arr)):
        key = arr[i] 
        # Move elements of arr[0..i-1], that are 
        # greater than key, to one position ahead 
        # of their current position 
        j = i-1
        while j >= 0 and key < arr[j] : 
                arr[j + 1] = arr[j] 
                j -= 1
        arr[j + 1] = key
        
# Driver code to test above 
arr = [12, 11, 13, 5, 6] 
insertionSort(arr) 
for i in range(len(arr)): 
    print ("% d" % arr[i]) 
```

```python
# Python program for implementation of MergeSort 
def mergeSort(arr): 
    if len(arr) >1: 
        mid = len(arr)//2 #Finding the mid of the array 
        L = arr[:mid] # Dividing the array elements  
        R = arr[mid:] # into 2 halves 
  
        mergeSort(L) # Sorting the first half 
        mergeSort(R) # Sorting the second half 
  
        i = j = k = 0
          
        # Copy data to temp arrays L[] and R[] 
        while i < len(L) and j < len(R): 
            if L[i] < R[j]: 
                arr[k] = L[i] 
                i+=1
            else: 
                arr[k] = R[j] 
                j+=1
            k+=1
          
        # Checking if any element was left 
        while i < len(L): 
            arr[k] = L[i] 
            i+=1
            k+=1
          
        while j < len(R): 
            arr[k] = R[j] 
            j+=1
            k+=1
  
# Code to print the list 
def printList(arr): 
    for i in range(len(arr)):         
        print(arr[i],end=" ") 
    print() 
  
# driver code to test the above code 
if __name__ == '__main__': 
    arr = [12, 11, 13, 5, 6, 7]  
    print ("Given array is", end="\n")  
    printList(arr) 
    mergeSort(arr) 
    print("Sorted array is: ", end="\n") 
    printList(arr) 
  
```

```python
# Given a list 'L' of sorted integers and another non-negative integer k, 
# find if there exist 2 indices i and j such that L[i] - L[j] = k, i != j.

# Reading the input
import ast 
L = ast.literal_eval(input())
k = int(input())

# Write your code here
left=0
right=len(L)-1
count=0
while left<right:
    if (L[right] - L[left])==k:
        count=count+1
        if count==2:
            print ("Yes")
            break
    
    elif (L[right] - L[left])!=k:
        left=left+1
    else:
        right=right-1
        
if count==0:
    print("No")
```

```python
# Earlier you found factorial using loops. 
# Here you have to apply the theory of recursive functions to find the factorial of a positive integer.

#take input here
num=int(input())

#write the recurssive function to find the factorial of num
def factorial(n):
    #find and RETURN factorial of num
    if n ==0:
        return 1
    else:
        prev = factorial(n-1)
        final_answer = n*prev
        return final_answer
    

#function ends here

print(factorial(num))
```

```python
# by writing a recursive function. You will be given a positive integer n you have to find the nth Fibonacci number.

n = int(input())

def fib(n):
    if n ==1 or n==2:
        return 1
    else:
        prev1 = fib(n-1)
        prev2 = fib(n-2)
        fib_val = prev1+prev2
        return fib_val
    
print(fib(n))
```
